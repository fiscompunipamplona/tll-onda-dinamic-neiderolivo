interferencia entre dos ondas en un estanque debido a dos fuentes de diferenetes longitud de onda y ubicacion 

```python
import numpy as np
info=np.zeros((200,200))
np.savetxt("array",info,fmt="%s")
```

```python
from pylab import imshow, show, xlabel, ylabel,plot,xlim,ylim,hot, magma
from numpy import loadtxt,linspace,sin,sqrt,cos
import numpy as np
from IPython import display as disp
import matplotlib.pyplot as plt
salida2=open("ondas_estanque.dat","w")
s=800
d=100
x=linspace(0,d,s)
y=linspace(0,d,s)
X,Y=np.meshgrid(x,y)
c1=1 #amplitud
c2=2 #aplitud particula 2
w1=2
w2=5#frecuencia
b1=5 #longitud de onda
b2=3
k1=(2*3.1416)/b1
k2=(2*3.1416)/b2
x1=25 #psicion inicial
y1=25 #posicion inicial y
x2=75 #posisicion inicial paticula 2
y2=55 #posisicion inicial y
t=20 #tiempo
info=np.zeros((s,s,t),dtype=float)
dimx=100
nx=500
dx=dimx/nx
for t in range(0,t):
    r1=sqrt((X-x1)**2 +(Y-y1)**2)
    r2=sqrt((X-x2)**2 +(Y-y2)**2)
    C=c1*sin(k1*r1+w1*t)+c2*sin(k2*r2+w2*t)
    plt.gca().cla()
    info[:,:,t]=C
    imshow(C,origin="lower", extent=[0, d, 0, d])
    magma()
    disp.clear_output(wait=True)
    disp.display(plt.gcf())
```

```python
for r in range (t):
    salida2.write('#matriz %d \n' %(r))
    for i in range(s):
        for j in range(s):
            salida2.write('%0.4f ' %(info[i,j,r]))
        salida2.write('\n')
salida2.close()
```
