 ondas en un estanque
<br><br>
- Suponga un estanque de agua en reposo. Represente las ondas generadas sobre el estanque cuando se deja caer una piedra en un cierto lugar.
<br>
- Teniendo que el punto donde cae la piedra está ubicado en ($x_1,y_1$), entonces la distancia $r_1$ desde donde se generan las ondas a otro punto ($x,y$) en el estanque es:
<br><br>
\begin{equation}
    r_1 = \sqrt{(x_1-x)^2+(y_1-y)^2}\, .
\end{equation}
<br><br>
- Asuma que la onda generada sigue una función seno:
\begin{equation}
    \zeta_1 = \zeta_0\sin(kr_1)\, ,
\end{equation}
<br><br>
- Donde $k$ representa el número de onda y la longitud de onda $\lambda = 5.0\,$cm y $\zeta_0 = 1.0\,$cm
<br><br>

```python
import numpy as np
info=np.zeros((200,200))
np.savetxt("array",info,fmt="%s")
```

```python
from pylab import imshow, show, xlabel, ylabel,plot,xlim,ylim,hot,jet,magma,plasma,inferno
from numpy import loadtxt,linspace,sin,sqrt
import numpy as np
from IPython import display as disp
import matplotlib.pyplot as plt
salida=open("ondas_estanque2.dat","w")
s=800 #diviciones 
d=50
x=linspace(0,d,s)
y=linspace(0,d,s)
X,Y=np.meshgrid(x,y)
t=20
info=np.zeros((s,s,t),dtype=float)
#xlim(0,500)
#ylim(0,500)
c=1 #amplitud
b=5 #longitud de onda
k=(2*3.1416)/b
x1=25 #psicion inicial
y1=25 #posicion inicial
W=100
dimx=100
nx=500
dx=dimx/nx
#salida=open("ondas_estanque2.dat","w")
for t in range(0,t):
    r1=sqrt((X-x1)**2 +(Y-y1)**2)
    C=c*sin(k*r1+W*t)
    plt.gca().cla()
    info[:,:,t]=C
    imshow(C,origin="lower", extent=[0, d, 0, d])
    magma()
    disp.clear_output(wait=True)
    disp.display(plt.gcf())
```

```python
for r in range (t):
    salida.write('#matriz %d \n' %(r))
    for i in range(s):
        for j in range(s):
            salida.write('%0.4f ' %(info[i,j,r]))
        salida.write('\n')
salida.close()
```
